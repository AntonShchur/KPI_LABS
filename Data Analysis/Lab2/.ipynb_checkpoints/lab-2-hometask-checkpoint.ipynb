```python
# Тепер імпортуємо Pandas:
import pandas as pd
```

2. Далі зчитаємо дані з файлу Data2.csv у data frame Pandas:

```python
DATA_PATH = 'data/Data2.csv'

# Зчитуємо файл із заданим розділяючим символом ";"
df = pd.read_csv(DATA_PATH, sep=';')

'''
Зауваження: попередній код зчитування з файлу підходить для
файлів з кодуванням utf-8. Проте файл Data2.csv напочатку мав інше
кодування. Дане кодування потрібно вказати в якості параметру:

df = pd.read_csv(DATA_PATH, sep=';', encoding='cp1252')

'''

print('Data frame info:')
df.info()

# Виводимо перші 5 рядків
print('\nFirst 5 rows:')
df.head()
```

3. Після дослідження структури даних, можна побачити, що формат чисел для стовпчиків "GDP per capita", "CO2 emission" та "Area" не підходить для приведення у тип ``float``, так як містить коми замість крапок. Також виведемо рядки з пропущеними даними:

```python
print('Columns with NA/NaN values:')
print(df.isna().any())

rows_with_missing_values = df.isna().any(axis=1)

print('\nRows with missing values:\n')
df[rows_with_missing_values]
```

4. Виправимо формат дробних чисел та приведемо їх до типу ``float``:

```python
def convert_column_to_float(dataset, column_label):
    dataset[column_label] = dataset[column_label] \
        .str.replace(',', '.') \
        .astype(float)

convert_column_to_float(df, 'GDP per capita')
convert_column_to_float(df, 'CO2 emission')
convert_column_to_float(df, 'Area')
```

5. Додатково перевіримо числові значення на адекватність. Значення ВВП, викидів CO2, площі та кількості населення не можуть бути від'ємними. Перевіримо, чи відповідають дані цьому факту:

```python
# Виводимо рядки з від'ємними значеннями по відповідним стовпчикам:
print('GDP per capita negative values:\n', df[df['GDP per capita'] < 0])
print('\nCO2 emission negative values:\n', df[df['CO2 emission'] < 0])
print('\nPopulation negative values:\n', df[df['Population'] < 0])
print('\nArea negative values:\n', df[df['Area'] < 0])
```

6. Як можна було побачити у попередньому пункті, деякі значення в наборі даних є некоректними. Виправимо дані некоректності:

```python
# Для виправлення некоректності зі знаками числових значень
# просто домножимо від'ємні на (-1).
gpd_rows_to_fix = df[df['GDP per capita'] < 0]
area_rows_to_fix = df[df['Area'] < 0]

gpd_rows_to_fix['GDP per capita'] *= -1
area_rows_to_fix['Area'] *= -1

df[df['GDP per capita'] < 0] = gpd_rows_to_fix
df[df['Area'] < 0] = area_rows_to_fix

# Виводимо рядки з від'ємними значеннями по відповідним стовпчикам:
print('GDP per capita negative values:\n', df[df['GDP per capita'] < 0])
print('\nCO2 emission negative values:\n', df[df['CO2 emission'] < 0])
print('\nPopulation negative values:\n', df[df['Population'] < 0])
print('\nArea negative values:\n', df[df['Area'] < 0])
```

7. Заповнимо пропущені дані середніми значеннями колонок:

```python
df = df.fillna(df.mean())

df.info()

print('\nColumns with NA/NaN values:')
print(df.isna().any())

# Виведемо рядки, які раніше містили пропущені значення
print('\nRows with previously missing values:')
df[rows_with_missing_values]
```

## Відповіді на питання:

1. Пропущені значення були присутні, рядки з пропущеними значеннями були показані у п.3 та замінені на середні значення відповідних стопвчиків у п.5 ходу роботи.
2. Знайдемо країни з найбільшим ВВП на людину та з найменшою площею:

```python
# Виведення інформації про країну з найбільшим ВВП на людину:
max_gdp_row = df.loc[df['GDP per capita'].idxmax()]
max_gdp_country_name = max_gdp_row['Country Name']
print('Country with a max. GDP per capita:', max_gdp_country_name)
print(f'\nInfo about {max_gdp_country_name}:')
print(max_gdp_row)

# Виведення інформації про країну з найменшою площею:
min_area_row = df.loc[df['Area'].idxmin()]
min_area_country_name = min_area_row['Country Name']
print('\nCountry with a min. area:', min_area_country_name)
print(f'\nInfo about {min_area_country_name}:')
print(min_area_row)
```

3. Знайдемо регіон, в якому середня площа країни найбільша:

```python
# Виведемо середні площі країн по регіонам
region_mean_areas = df.groupby(['Region']).mean()['Area']
print('Countries\' mean areas by regions:\n', region_mean_areas)

# Виведемо шуканий регіон
print('\nRegion with the biggest country mean area:', region_mean_areas.idxmax())
```

4. Знайдемо населення, з яким найчастіше зустрічаються країни у світі та у Європі зокрема:

```python
world_max_population_idx = df['Population'].idxmax()
print('Country with the biggest population in the world:')
print(df.loc[world_max_population_idx][['Country Name', 'Population']])

europe_max_population_idx = df[df['Region'] == 'Europe & Central Asia']['Population'].idxmax()
print('\nCountry with the biggest population in Europe & Central Asia:')
print(df.loc[europe_max_population_idx][['Country Name', 'Population']])
```

5. З'ясуємо, чи співпадає в якомусь із регіонів середнє значення та медіана ВВП:

```python
# Розраховуємо та виводимо середні значення і медіани ВВП країн по регіонам:
region_mean_gdp = df.groupby(['Region']).mean()['GDP per capita']
region_median_gdp = df.groupby(['Region']).median()['GDP per capita']

print('Countries\' mean GDP per capita by region:\n', region_mean_gdp)
print('\nCountries\' median GDP per capita by region:\n', region_median_gdp)

# Знаходимо перетин середніх значень і медіан. Як бачимо, результат не має елементів,
# отже, регіонів із співпадаючим середнім значенням та медіаною ВВП немає. Проте,
# доволі близьким є регіон "Latin America & Caribbean".
print(pd.merge(region_mean_gdp, region_median_gdp, how='inner'))
```

6. Виведемо топ 5 та 5 останніх країн по ВВП та кількості викидів CO2 на душу населення:

```python
print('Top 5 countries by GDP per capita:')
df.sort_values(by=['GDP per capita'], ascending=False).head()
```

```python
print('\nLowest 5 countries by GDP per capita:')
df.sort_values(by=['GDP per capita']).head()
```

```python
# Додамо стовпчик з викидами вуглекислого газу на особу
df['CO2 emission per citizen'] = df['CO2 emission'] / df['Population']

print('\nTop 5 countries by CO2 emmisions per citizen:')
df.sort_values(by=['CO2 emission per citizen'], ascending=False).head()
```

```python
print('\nLowest 5 countries by CO2 emmisions per citizen:')
df.sort_values(by=['CO2 emission per citizen']).head()
```
